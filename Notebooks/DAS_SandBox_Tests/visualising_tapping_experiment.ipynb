```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import numpy as np
import os


import matplotlib.pyplot as plt
```

```python
datapath = '/media/birniece/Extreme SSD/HydraFrac/SandBox_DAS_data/Processed data/Processed data'
```

```python
def DAS_to_2D(origdata):
    data_3D = origdata.reshape([origdata.shape[0],origdata.shape[1]//3,3])
    data_3D = data_3D.transpose([2,0,1])
    data_2D = data_3D.reshape([3,data_3D.shape[1]*data_3D.shape[2]])
    return data_2D
```

```python
w1_data = DAS_to_2D(np.loadtxt(os.path.join(datapath,'Wrap_1_time.csv'), delimiter=','))
w2_data = DAS_to_2D(np.loadtxt(os.path.join(datapath,'Wrap_2_time.csv'), delimiter=','))
w3_data = DAS_to_2D(np.loadtxt(os.path.join(datapath,'Wrap_3_time.csv'), delimiter=','))
```

```python
t = np.arange(len(w1_data[1]))

fig,axs = plt.subplots(3,1,figsize=[12,5],sharex=True, sharey=True)

axs[0].plot(t,w1_data[1], 'k')
axs[1].plot(t,w2_data[1], 'k')
axs[2].plot(t,w3_data[1], 'k')

axs[0].set_ylabel('Layer 1',fontsize=14)
axs[1].set_ylabel('Layer 2',fontsize=14)
axs[2].set_ylabel('Layer 3',fontsize=14)
axs[2].set_xlabel('Time Samples',fontsize=14)
axs[2].set_xlim([0,max(t)])
plt.suptitle('Tap Test - Time Domain',fontsize=18)
plt.tight_layout()
plt.savefig('./TapTest_TimeDomain.png')
```
