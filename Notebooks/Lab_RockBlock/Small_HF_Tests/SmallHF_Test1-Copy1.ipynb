```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
trans_file = '/home/birniece/Documents/data/ANPERC_Lab/LargeRockBlock/SmallHFexps/HF_small_4.csv'
```

```python
df = pd.read_csv(trans_file, skiprows=9, index_col=0, header=None )
df.dropna(axis=1, inplace=True)
df.columns = ['R1','R2','R3','R4']
df.head()
```

```python
data = np.vstack([df['R1'].values, 
                  df['R2'].values,
                  df['R3'].values,
                  df['R4'].values,
                 ])
t = df.index.values

# Demean
data_dmn = data - np.tile(np.expand_dims(np.mean(data, axis=1),1),(1,data.shape[1]))
```

```python
fig,axs = plt.subplots(4,1,figsize=[12,6], sharex=True)
for i in range(4):
    axs[i].plot(t, data_dmn[i])
    axs[i].set_title('R%i'%(i+1))
    
axs[-1].set_xlabel('Time (s)')
axs[-1].set_xlim([0,220])
    
fig.suptitle('Small HF Test 4')
fig.tight_layout()
```
