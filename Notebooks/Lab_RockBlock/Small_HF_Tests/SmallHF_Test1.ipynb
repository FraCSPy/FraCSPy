```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
trans_file = '/home/birniece/Documents/data/ANPERC_Lab/LargeRockBlock/SmallHFexps/HF-small_1.csv'
```

```python
df = pd.read_csv(trans_file, skiprows=9, index_col=0, header=None )
df.dropna(axis=1, inplace=True)
df.columns = ['R1','R2','R3','R4']
df.head()
```

```python
fig,axs = plt.subplots(4,1,figsize=[12,6], sharex=True)
axs[0].plot(df['R1'])
axs[1].plot(df['R2'])
axs[2].plot(df['R3'])
axs[3].plot(df['R4'])

axs[0].set_title('R1')
axs[1].set_title('R2')
axs[2].set_title('R3')
axs[3].set_title('R4')

axs[-1].set_xlabel('Time (s)')
# axs[-1].set_xlim([0,110])
    
fig.suptitle('Small HF Test 1')
fig.tight_layout()
```
