```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pyfrac.visualisation.traceviz import wiggleplot
```

```python
sty_file = '/home/birniece/Documents/Projects/HydraFrac/Stylolite_Investigation/STY_all_1.xlsx'
r = []
for i in range(1,7):
    df = pd.read_excel(sty_file,sheet_name=i, names=['time','amp'])
    r.append(df['amp'].values[:1000])
    
rec_data = np.vstack(r)
fig,axs = wiggleplot(rec_data[:,:50]);
fig.suptitle('STYLOLITES')

plt.figure(figsize=[15,3])
plt.plot(rec_data.T);

plt.figure(figsize=[15,3])
plt.plot(rec_data[:,:50].T);
```

```python
sty_file = '/home/birniece/Documents/Projects/HydraFrac/Stylolite_Investigation/STY_ALL.xlsx'
for i in range(1,8):
    print(i)
    if i ==1:    
        df = pd.read_excel(sty_file,sheet_name=i, names=['time',i])
    else:
        df_tmp = pd.read_excel(sty_file,sheet_name=i, names=['time',i])
        df = df.merge(df_tmp, on='time')
        
# Set time as index
df.set_index('time', inplace=True)
rec_data = df.values
time_data = df.index.values

fig,axs = wiggleplot(rec_data[200:750].T);
fig.suptitle('STYLOLITES')

plt.figure(figsize=[15,3])
plt.plot(time_data, rec_data);

plt.figure(figsize=[15,3])
plt.plot(time_data[100:750], rec_data[100:750]);
```
