```python
import glob
import obspy

# Import necessary libraries
# glob: For finding pathnames matching a specified pattern
# obspy: For working with seismic data

# Read Certain Event
# Define the path to the folder containing waveform files for certain events
path_event = glob.glob('../../data/TOC2ME_Data/waveforms/*')

# Read the Third Event from the Folder
# Select the third event (index 2 in the list)
k = 2
path_data = glob.glob(path_event[k] + '/*')

# Append data from different stations
# Read and merge data from different stations for the selected event
st = obspy.read(path_data[0])

# Iterate through the remaining files in the folder and append their data
for i in range(1, len(path_data)):
    st += obspy.read(path_data[i])

# Merge the data from different stations (optional)
# Note: You may want to uncomment this line if you intend to merge the data
# st = st.merge()

# Plot the merged seismic data
st.plot()

# Print summary information about the merged stream
print(st.__str__(extended=True))
```
