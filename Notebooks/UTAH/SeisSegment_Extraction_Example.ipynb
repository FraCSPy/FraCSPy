## Extract Data Segments from Original Recorded Seismograms

```python
import obspy
from obspy.core import UTCDateTime
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Function - extracting data segments directly from the .fcnt file
def extract_segment(filename, start_time, end_time):
    segment = obspy.read(filename, format='rg16', merge=True, starttime=start_time, endtime=end_time)
    return segment
```

```python
ref_time = UTCDateTime("2019-04-27T17:29:49")   # reference time for a detected event

# extract a data segment 0.2 second before and 1.0 second after the reference time
start_time = ref_time - 0.2
end_time   = ref_time + 1.0
```

```python
filename = f"/home/birniece/Documents/data/UTAH_FORGE/Ap2019_SurfaceNodes/1.fcnt"
segment_3C = extract_segment(filename, start_time, end_time)
```

```python
# to show the information of the extracted segment
print(segment_3C)
```

```python
# to see the recorded data of the first component in the extarced segment
print(segment_3C[0].data)
```
